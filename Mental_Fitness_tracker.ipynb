## Analyzing the data

### Loading the data

```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/prevalence-by-mental-and-substance-use-disorder.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mental-and-substance-use-as-share-of-disease.csv')
```

```python
# mental-and-substance-use-as-share-of-disease
df2.head()
```

```python
# Merging both the data
dataset = pd.merge(df1,df2)
dataset.head()
```

```python
# Changing the columns name
dataset.set_axis(['Country','Code','Year','Schizophrenia','Bipolar_Disorder','Eating_Disorder','Anxiety_Disorder','Drug_Use_Disorder','Depressive_Disorder','Alcohol_Use_Disorder','Mental_Disorder'],axis='columns',inplace=True)
dataset.head()
```

### Differnt visualizations for the dataset

```python
fig = px.pie(dataset, values='Mental_Disorder', names='Year')
fig.show()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl=LabelEncoder()
for i in dataset.columns:
  if dataset[i].dtype=="object":
    dataset[i]=lbl.fit_transform(dataset[i])
```

```python
dataset.shape
```

### Splitting the data

```python
x = dataset.drop('Mental_Disorder',axis=1)
y = dataset['Mental_Disorder']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=.05,random_state=1)
```

```python
print('xtrain = ',xtrain.shape)
print('xtest = ',xtest.shape)
print('ytrain = ',ytrain.shape)
print('ytest = ',ytest.shape)
```

## Training the model

### Random Forest Regressor
